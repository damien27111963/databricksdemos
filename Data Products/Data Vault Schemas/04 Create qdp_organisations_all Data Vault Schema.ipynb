```sql
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_organisation_addresses AS 
SELECT 
  h.tennant_id,
  t.tennant_name,
  s.organisation_name,
  sadd.full_address,
  h.organisation_id,
	h.organisation_entity_id,
	laddhist.is_primary_address,
  laddhist.use_code AS address_use_code,
  sadd.postal_code,
  laddhist.period_start AS resided_at_address_from,
  laddhist.period_end AS resided_at_address_until,
	h.period_start AS ind_hub_period_start,
	h.period_end AS ind_hub_period_end,
  s.sat_organisation_id,


	s.period_start AS ind_sat_period_start,
	s.period_end AS ind_sat_period_end

    FROM demo_banking_silver.qdp_organisations_all.hub_organisation h
    JOIN demo_banking_silver.qdp_organisations_all.sat_organisation s
      ON h.organisation_id = s.organisation_id
    JOIN demo_banking_silver.qdp_links_all.link_organisation_address_history laddhist
      ON h.organisation_id = laddhist.organisation_id
    JOIN demo_banking_silver.qdp_locations_all.hub_address hadd
      ON laddhist.address_id = hadd.address_id
    JOIN demo_banking_silver.qdp_locations_all.sat_address sadd
      ON hadd.address_id = sadd.address_id
    JOIN qdp_refcode_silver.qdp_reference_codes.tennant t
      on h.tennant_id = t.tennant_id
    ORDER BY h.organisation_id ASC
;

SELECT * from view_organisation_addresses;
```

```sql
%sql

USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_organisation_risk_scores AS 
SELECT 
  h.organisation_id, 
  h.organisation_entity_id, 
  h.tennant_id, 
  t.tennant_name,
  s.organisation_name, 
  s.company_registration_country_code, 
  r.risk_code, 
  r.risk_raw_code, 
  r.risk_score 
     FROM demo_banking_silver.qdp_organisations_all.hub_organisation h
     JOIN demo_banking_silver.qdp_organisations_all.sat_organisation s
       ON h.organisation_id = s.organisation_id
     JOIN demo_banking_silver.qdp_organisations_all.sat_organisation_risk_scores r   
       ON s.organisation_id = r.organisation_id
     JOIN qdp_refcode_silver.qdp_reference_codes.tennant t
      ON h.tennant_id = t.tennant_id
  
--   WHERE r.risk_score > 50
   ORDER BY s.organisation_name, r.risk_score
;


SELECT * FROM  view_organisation_risk_scores;
```

```sql
%sql

USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_organisations_with_high_risk_scores AS 
SELECT 
  h.organisation_id, 
  h.organisation_entity_id, 
  h.tennant_id, 
  t.tennant_name,
  s.organisation_name, 
  s.company_registration_country_code, 
  r.risk_code, 
  r.risk_raw_code, 
  r.risk_score 
     FROM demo_banking_silver.qdp_organisations_all.hub_organisation h
     JOIN demo_banking_silver.qdp_organisations_all.sat_organisation s
       ON h.organisation_id = s.organisation_id
     JOIN demo_banking_silver.qdp_organisations_all.sat_organisation_risk_scores r   
       ON s.organisation_id = r.organisation_id
     JOIN qdp_refcode_silver.qdp_reference_codes.tennant t
      ON h.tennant_id = t.tennant_id
  
   WHERE r.risk_score > 50
   ORDER BY s.organisation_name, r.risk_score
;


SELECT * FROM  view_organisations_with_high_risk_scores;
```
