```sql

--SELECT version();

CREATE SCHEMA IF NOT EXISTS qdp_refcodes_all;
SET search_path TO qdp_refcodes_all;


-- metadata table
DROP TABLE IF EXISTS metadata;


CREATE TABLE metadata ( 
	supplier VARCHAR,
	data_product VARCHAR,
	data_product_version VARCHAR,
	schema_name VARCHAR,
	schema_version VARCHAR,
	schema_variant_name VARCHAR,
	schema_variant_version VARCHAR,
	instance VARCHAR,
	instance_version VARCHAR
 );



INSERT INTO metadata 
(
  supplier, 
  data_product,
  data_product_version,
  schema_name,
  schema_version,
  schema_variant_name,
  schema_variant_version,
  instance,
  instance_version
) 
VALUES (
  'quantexa.com',
  'com.quantexa.qdp.refcodes_all',
  '0.1',
  'qdp_refcodes_all',
  '0.1',
  'universal',
  '0.1',
  'not yet populated with data',
  ''
);




-- tennant table
DROP TABLE IF EXISTS tennant;

CREATE TABLE tennant ( 
	tennant_id BIGINT,
	tennant_name VARCHAR,
	network_name VARCHAR
 );

INSERT INTO tennant 
(
  tennant_id, 
  tennant_name,
  network_name
) 
VALUES (
  0,
  'NO TENNANT',
  'NO NETWORK'
);

INSERT INTO tennant 
(
  tennant_id, 
  tennant_name,
  network_name
) 
VALUES (
  1,
  'Banking Supply Chain Demo',
  'CS Post-It Out LTD - Trade & Working Capital'
);

INSERT INTO tennant 
(
  tennant_id, 
  tennant_name,
  network_name
) 
VALUES (
  2,
  'FinCrime Demo',
  'Nine Gee LTD'
);



-- Record Source table
DROP TABLE IF EXISTS record_source;

CREATE TABLE record_source ( 
	record_source_id BIGINT,
	domain VARCHAR,
	it_system VARCHAR,
	network_name VARCHAR
 );

INSERT INTO record_source 
(
  record_source_id, 
  domain,
  it_system,
  network_name
) 
VALUES (
  0,
  'NOT DEFINED',
  'NOT DEFINED',
  'NOT DEFINED'
);


INSERT INTO record_source 
(
  record_source_id, 
  domain,
  it_system,
  network_name
) 
VALUES (
  1,
  'com.quantexa.demo/banking',
  '',
  'CS Post-It Out LTD - Trade & Working Capital'
);


INSERT INTO record_source 
(
  record_source_id, 
  domain,
  it_system,
  network_name
) 
VALUES (
  2,
  'com.quantexa.demo/banking',
  '',
  'Nine Gee LTD'
);




-- 

-- Create the "Before Quantexa Processing Demo" Tennant
INSERT INTO tennant 
(
  tennant_id, 
  tennant_name,
  network_name
) 
VALUES (
  999,
  'Before Quantexa Processing Demo',
  'CS Post-It Out LTD - Trade & Working Capital'
);

INSERT INTO record_source 
(
  record_source_id, 
  domain,
  it_system,
  network_name
) 
VALUES (
  999,
  'com.quantexa.demo/banking',
  '',
  'CS Post-It Out LTD - Trade & Working Capital'
);


-- SELECT * from tennant;


--
--  Vocabulary Concept Tables
--

-- Domain table
DROP TABLE IF EXISTS Domain;

CREATE TABLE domain (
	domainId VARCHAR NOT NULL,
	domainConceptId BIGINT NOT NULL,
	domainName VARCHAR NOT NULL
);

ALTER TABLE Domain ADD CONSTRAINT pk_conceptAll_domain_domainId PRIMARY KEY (domainId);



-- Vocabulary table
DROP TABLE IF EXISTS Vocabulary;

CREATE TABLE Vocabulary (
	vocabularyId VARCHAR NOT NULL,
	vocabularyConceptId BIGINT NOT NULL,
	vocabularyName VARCHAR NOT NULL,
	vocabularyReference VARCHAR NOT NULL,
	vocabularyVersion VARCHAR NOT NULL
);

ALTER TABLE Vocabulary ADD CONSTRAINT pk_conceptAll_vocabulary_vocabularyId PRIMARY KEY (vocabularyId);



-- ConceptClass table
DROP TABLE IF EXISTS ConceptClass;

CREATE TABLE ConceptClass (
	conceptClassId VARCHAR NOT NULL,
	conceptClassConceptId BIGINT NOT NULL,
	conceptClassName VARCHAR NOT NULL
);

ALTER TABLE ConceptClass ADD CONSTRAINT pk_conceptAll_conceptClass_conceptClassId PRIMARY KEY (conceptClassId);

-- Concept table
DROP TABLE IF EXISTS Concept;

CREATE TABLE Concept (
	conceptId BIGINT NOT NULL,
	domainId VARCHAR NOT NULL,
	vocabularyId VARCHAR NOT NULL,
	conceptClassId VARCHAR NOT NULL,
	standardConcept VARCHAR NOT NULL,
	conceptCode VARCHAR NOT NULL,
  validStartDate DATE NOT NULL,
  validEndDate DATE NOT NULL,
  invalidReason VARCHAR
);

ALTER TABLE Concept ADD CONSTRAINT pk_conceptAll_concept_conceptId PRIMARY KEY (conceptId);


-- ConceptSynonym table
DROP TABLE IF EXISTS ConceptSynonym;

CREATE TABLE ConceptSynonym (
	conceptId BIGINT NOT NULL,
	languageConceptId BIGINT NOT NULL,
	conceptSynonymName VARCHAR NOT NULL
);


-- Relationship table
DROP TABLE IF EXISTS Relationship;

CREATE TABLE Relationship (
	relationshipId VARCHAR NOT NULL,
	relationshipConceptId BIGINT NOT NULL,
  reverseRelationshipId VARCHAR NOT NULL,
	relationshipName VARCHAR NOT NULL,
  isHierarchical BOOLEAN NOT NULL DEFAULT FALSE,
  definesAncestry BOOLEAN NOT NULL DEFAULT FALSE
);

ALTER TABLE Relationship ADD CONSTRAINT pk_conceptAll_relationship_relationshipId PRIMARY KEY (relationshipId);


-- Concpet Relationship table
DROP TABLE IF EXISTS ConceptRelationship;

CREATE TABLE ConceptRelationship (
	conceptId1 BIGINT NOT NULL,
	conceptId2 BIGINT NOT NULL,
	relationshipId VARCHAR NOT NULL,
  validStartDate DATE NOT NULL,
  validEndDate DATE NOT NULL,
  invalidReason VARCHAR
);

ALTER TABLE ConceptRelationship ADD CONSTRAINT pk_conceptAll_ConceptRelationship_conceptId1_conceptID2 PRIMARY KEY (conceptId1, conceptId2);


```
