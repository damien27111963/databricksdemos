```sql
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_banking_supply_chains AS 
SELECT 
  h.tennant_id as tennant_id,
	h.tennant_name AS tennant_name,
  h.network_name AS network_name,
  beneacc.account_name as beneficiary_account_name,
  beneacc.account_number as beneficiary_account_number,
  cast(beneacc.is_customer_account as STRING) as is_beneficiary_customer,
  origacc.account_name as originator_account_name,
  origacc.account_number as originator_account_number,
  cast(origacc.is_customer_account as STRING) as is_originator_customer,
  s.total_value,
	s.max_value,
	s.min_value,
	s.avg_value,
  s.first_transaction_date,
	s.last_transaction_date,
  beneacc.country_code as beneficiary_account_country,
  origacc.country_code as originator_account_country,
  h.banking_supply_chain_id,
  h.bene_account_id,
  h.bene_account_entity_id,
  h.orig_account_id,
  h.orig_account_entity_id,
  h.period_start as hub_period_start,
  h.period_end as hub_period_end,
  s.period_start,
	s.period_end,
	s.load_datetime,
	s.record_source_id,
	s.type_code,
	s.type_concept_id,
	s.type_raw_code,
	s.type_raw_concept_id,
	s.data_source_code,
	s.data_source_concept_id,
	s.data_source_raw_code,
	s.data_source_raw_concept_id


    FROM demo_banking_gold.qdp_banking_supply_chain.hub_banking_supply_chain h
    JOIN demo_banking_gold.qdp_banking_supply_chain.sat_banking_supply_chain s
      ON h.banking_supply_chain_id = s.banking_supply_chain_id
    JOIN demo_banking_silver.qdp_accounts_all.sat_account beneacc
      ON beneacc.account_id = h.bene_account_id
    JOIN demo_banking_silver.qdp_accounts_all.sat_account origacc
      ON origacc.account_id = h.orig_account_id
    JOIN demo_banking_silver.qdp_refcodes_all.tennant t
      ON h.tennant_id = t.tennant_id
    ORDER BY beneacc.account_name 
;

SELECT * from view_banking_supply_chains;
```

```sql
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_banking_supply_chain_transactions AS 
SELECT 
  h.tennant_id as tennant_id,
	h.tennant_name AS tennant_name,
  h.network_name AS network_name,
  beneacc.account_name as beneficiary_account_name,
  beneacc.account_number as beneficiary_account_number,
  beneacc.country_code as beneficiary_account_country,
  CAST(beneacc.is_customer_account as STRING) as is_beneficiary_customer,
  origacc.account_name as originator_account_name,
  origacc.account_number as originator_account_number,
  origacc.country_code as originator_account_country,
  CAST(origacc.is_customer_account as STRING) as is_originator_customer,
  s.total_value,
	s.max_value,
	s.min_value,
	s.avg_value,
  s.first_transaction_date,
	s.last_transaction_date,
  strans.datetime,
  strans.amount,
  strans.debit_or_credit_code,
  strans.original_currency_code,
  strans.currency_code,
  strans.is_international_transaction


    FROM demo_banking_gold.qdp_banking_supply_chain.hub_banking_supply_chain h
    JOIN demo_banking_gold.qdp_banking_supply_chain.sat_banking_supply_chain s
      ON h.banking_supply_chain_id = s.banking_supply_chain_id
    JOIN demo_banking_silver.qdp_accounts_all.sat_account beneacc
      ON beneacc.account_id = h.bene_account_id
    JOIN demo_banking_silver.qdp_accounts_all.sat_account origacc
      ON origacc.account_id = h.orig_account_id
    JOIN demo_banking_silver.qdp_transactions_all.hub_transaction htrans
      ON beneacc.account_id = htrans.bene_account_id AND origacc.account_id = htrans.orig_account_id
    JOIN demo_banking_silver.qdp_transactions_all.sat_transaction strans
      ON htrans.transaction_id = strans.transaction_id    
    JOIN demo_banking_silver.qdp_refcodes_all.tennant t
      ON h.tennant_id = t.tennant_id

    ORDER BY h.banking_supply_chain_id
;

SELECT * from view_banking_supply_chain_transactions;
```
