```sql
-- Create Quantexa Data Product individual_customers Schema
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

-- Organisation Customer table
DROP TABLE IF EXISTS hub_product;

CREATE OR REPLACE TABLE hub_product (
    product_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1) COMMENT 'Product Identity',
    entity_id STRING,
  	tennant_id BIGINT NOT NULL DEFAULT 0,
		period_start TIMESTAMP,
		period_end TIMESTAMP
) USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported');

ALTER TABLE hub_product ADD CONSTRAINT pk_products_hubproduct_productid PRIMARY KEY (product_id);
```

```sql
-- Create Quantexa Data Product individual_customers Schema
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

-- Individual Customer table
DROP TABLE IF EXISTS sat_product;

CREATE OR REPLACE TABLE sat_product (
    sat_product_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1) COMMENT 'Product Identity Sat',
    product_id BIGINT NOT NULL COMMENT 'Product Identity',
    load_datetime TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP(),
    record_source_id BIGINT NOT NULL DEFAULT 1,
		product_name STRING,
		product_description STRING,
		product_code STRING,
		product_concept_id BIGINT,
		product_raw_code STRING,
		product_raw_concept_id BIGINT,
		maxPersonsAllowed INT,
		monthlyCost DECIMAL(10,2),
	  currency_code STRING,
	  currency_concept_id BIGINT,
	  currency_raw_code STRING,
	  currency_raw_concept_id BIGINT,
	  type_code STRING,
	  type_concept_id BIGINT,
	  type_raw_code STRING,
	  type_raw_concept_id BIGINT,
		period_start TIMESTAMP,
		period_end TIMESTAMP
) USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported');

ALTER TABLE sat_product ADD CONSTRAINT pk_products_satproduct_satproductid PRIMARY KEY (sat_product_id);
ALTER TABLE sat_product ADD CONSTRAINT fk_products_satproduct_hubproduct_productid FOREIGN KEY (product_id) REFERENCES hub_product(product_id);
```
