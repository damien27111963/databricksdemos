```sql
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_individual_contact_details AS 
SELECT 
  h.individual_id,
	h.individual_entity_id,
  h.tennant_id,
  t.tennant_name,
	ihn.is_trusted AS is_trusted_name,
  ihn.given1 AS given_name,
  ihn.family AS family_name,
  icm.rank AS communication_method_rank,
  icm.is_primary_communication_method,
  icm.value AS contact_value,
  icm.type_code AS contact_type_code,
  icm.use_code AS contact_use_code,
  icm.telephony_country_code,
  icm.period_start AS contact_method_start,
  icm.period_end AS contact_method_end,
  laddhist.address_id,
  laddhist.is_primary_address,
  laddhist.use_code AS address_use_code,
  sadd.full_address,
  sadd.postal_code,
  laddhist.period_start AS resided_at_address_from,
  laddhist.period_end AS resided_at_address_until,
	h.national_id,
	h.period_start AS ind_hub_period_start,
	h.period_end AS ind_hub_period_end,
  s.sat_individual_id,
	s.birth_date,
	s.birth_place_city,
	s.birth_place_country,
	s.is_deceased,
	s.deceased_datetime,
	s.gender_code,
	s.marital_status_code,
	s.ethnicity_code,
	s.religion_code,
	s.occupation_code,
	s.nationality_code,
	s.country_of_residence_code,
	s.is_employee,
	s.is_ex_employee,
	s.is_graduate,
	s.is_interpreter_required,
	s.is_employed,
	s.is_retired,
	s.is_student,
	s.is_individual_merged,
	s.period_start AS ind_sat_period_start,
	s.period_end AS ind_sat_period_end

    FROM demo_banking_silver.qdp_individuals_all.hub_individual h
    JOIN demo_banking_silver.qdp_individuals_all.sat_individual s
      ON h.individual_id = s.individual_id
    JOIN demo_banking_silver.qdp_individuals_all.sat_human_name ihn 
      ON h.individual_id = ihn.individual_id
    JOIN demo_banking_silver.qdp_individuals_all.sat_communication_method icm 
      ON h.individual_id = icm.individual_id
    JOIN demo_banking_silver.qdp_links_all.link_individual_address_history laddhist
      ON h.individual_id = laddhist.individual_id
    JOIN demo_banking_silver.qdp_locations_all.hub_address hadd
      ON laddhist.address_id = hadd.address_id
    JOIN demo_banking_silver.qdp_locations_all.sat_address sadd
      ON hadd.address_id = sadd.address_id
    JOIN qdp_refcode_silver.qdp_reference_codes.tennant t
      on h.tennant_id = t.tennant_id
    WHERE ihn.is_preferred 
    ORDER BY icm.rank
;

SELECT * from view_individual_contact_details;
```

```sql
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

CREATE OR REPLACE VIEW view_individual_communication_methods AS 
SELECT 
  h.individual_id,
	h.individual_entity_id,
  h.tennant_id,
  t.tennant_name,
	h.period_start AS hub_period_start,
	h.period_end AS hub_period_end,
  s.sat_individual_id,
	ocm.rank AS communication_method_rank,
	ocm.is_primary_communication_method AS communication_method_is_primary,
	ocm.value AS communication_method_value,
	ocm.value_display AS communication_method_value_display,
	ocm.value_raw AS communication_method_value_raw,
	ocm.type_code AS communication_method_type_code,
	ocm.type_concept_id AS communication_method_type_concept_id,
	ocm.type_raw_code AS communication_method_type_raw_code,
	ocm.type_raw_concept_id AS communication_method_type_raw_concept_id,
	ocm.use_code AS communication_method_use_code,
	ocm.use_concept_id AS communication_method_use_concept_id,
	ocm.use_raw_code AS communication_method_use_raw_code,
	ocm.use_raw_concept_id AS communication_method_use_raw_concept_id,
	ocm.telephony_country_code AS communication_method_telephony_country_code,
	ocm.telephony_country_concept_id AS communication_method_telephony_country_concept_id,
	ocm.telephony_country_raw_code AS communication_method_telephony_country_raw_code,
	ocm.telephony_country_raw_concept_id AS communication_method_telephony_country_raw_concept_id,
	ocm.period_start AS communication_method_period_start,
	ocm.period_end AS communication_method_period_end,
 	s.period_start,
	s.period_end,
	s.load_datetime,
	s.record_source_id,
	s.data_source_code,
	s.data_source_concept_id,
	s.data_source_raw_code,
	s.data_source_raw_concept_id

    FROM demo_banking_silver.qdp_individuals_all.hub_individual h
    JOIN demo_banking_silver.qdp_individuals_all.sat_individual s
      ON h.individual_id = s.individual_id
    JOIN demo_banking_silver.qdp_individuals_all.sat_communication_method ocm
      ON h.individual_id = ocm.individual_id
    JOIN qdp_refcode_silver.qdp_reference_codes.tennant t
      on h.tennant_id = t.tennant_id
;

SELECT * from view_individual_communication_methods;
```
