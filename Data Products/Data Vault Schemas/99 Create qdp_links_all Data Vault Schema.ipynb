```sql
-- create Link Table
USE CATALOG IDENTIFIER(:p_catalog_name);
USE SCHEMA IDENTIFIER(:p_schema_name);

DROP TABLE IF EXISTS link_organisation_address_history;

-- Organisation Address History
CREATE TABLE link_organisation_address_history ( 
	organisation_address_history_id BIGINT NOT NULL GENERATED  ALWAYS AS IDENTITY  (START WITH 1 INCREMENT BY 1) COMMENT 'Organisation Address History Identity',
	tennant_id BIGINT NOT NULL DEFAULT 0,
	organisation_id BIGINT NOT NULL,
	organisation_entity_id STRING NOT NULL,
	address_id BIGINT NOT NULL,
	address_entity_id STRING NOT NULL,
	is_primary_address BOOLEAN NOT NULL DEFAULT FALSE,
	use_code STRING,
	use_concept_id BIGINT,
	use_raw_code STRING,
	use_raw_concept_id BIGINT,
	period_start TIMESTAMP,
	period_end TIMESTAMP,
	CONSTRAINT pk_organisationall_organisationaddresshistory_organisationaddresshistoryid PRIMARY KEY ( organisation_address_history_id )
 )     USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported');


ALTER TABLE link_organisation_address_history ADD CONSTRAINT fk_qdplinks_organisationaddresshistory_huborganisation_organisationid FOREIGN KEY (organisation_id) REFERENCES demo_banking_silver.qdp_organisations_all.hub_organisation(organisation_id);

ALTER TABLE link_organisation_address_history ADD CONSTRAINT fk_qdplinks_organisationaddresshistory_hubaddress_addressid FOREIGN KEY (address_id) REFERENCES demo_banking_silver.qdp_locations_all.hub_address(address_id);
```
